```python
import requests
```

```python
laptop_url = 'https://www.amazon.in/s?k=laptop&crid=27GMOESJEPVWG&sprefix=laptop%2Caps%2C213&ref=nb_sb_noss_1'
response = requests.get(laptop_url)
```

```python
response.status_code
```

```python
page_content = response.text
page_content
```

```python
from bs4 import BeautifulSoup
```

```python
doc = BeautifulSoup(page_content, "html.parser")
doc
```

# What we want to find:
- Name of the laptop
- Processor
- RAM
- Memory

```python
selection_name = "a-size-medium a-color-base a-text-normal"
name_lap = doc.find_all('span', {'class' :selection_name})
name_lap
```

```python
s=name_lap[0].text.strip()
```

```python
s[:s.index(' ')]
```

```python
for tag in range(len(name_lap)):
    s=name_lap[tag]
    print(s)
```

```python
#def only_the_first_name(s):
name_list=[]
for tag in range(len(name_lap)):
    s=name_lap[tag].text.strip()
    name_list.append(s[:s.index(' ')])

name_list
```

```python
selection_price = "a-price-whole"
price = doc.find_all('span', {'class' : selection_price})
price
```

```python
price[0].text.replace(",",'')
```

```python
price_list = []
for tag in range(len(price)):
    price_list.append(int(price[tag].text.replace(",",'')))
price_list
    
```

```python
price_list_final=price_list[:3]+price_list[6:]
price_list_final
```

```python
selection_star1 = "a-icon a-icon-star-small a-star-small-4 aok-align-bottom"
star1 = doc.find_all('i',{'class' : selection_star1})
star1
```

```python
selection_star2 = 'a-icon a-icon-star-small a-star-small-3-5 aok-align-bottom'
star2 = doc.find_all('i',{'class' : selection_star2})
star2
```

```python
selection_star3 = 'a-icon a-icon-star-small a-star-small-2-5 aok-align-bottom'
star3 = doc.find_all('i',{'class' : selection_star3})
star3
```

```python
star=star1+star2+star3
len(star)
```

```python
v=star[0].text
```

```python
float(v[:v.index(' ')])
```

```python
star_list=[]
for tag in range(len(star)):
    v=star[tag].text
    star_list.append(float(v[:v.index(' ')]))

star_list
```

```python
len(name_list), len(price_list_final),len(star_list)
```

```python
import pandas as pd

all_data_dict ={
    'Name' :name_list,
    'Price' : price_list_final,
    'Star Rating' :star_list
}
```

```python
df_amazon = pd.DataFrame(all_data_dict)
df_amazon 
```
